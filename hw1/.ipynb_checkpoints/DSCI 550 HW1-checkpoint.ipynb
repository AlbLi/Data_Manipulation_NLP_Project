```python
# Function to retrieve affiliation based on citation source
def retrieve_affiliation(citation, doi):
    # Check the citation source and construct the URL accordingly
    if 'PLOS One' in citation:
        url = f"https://journals.plos.org/plosone/article?id={doi}"
    elif 'mBio' in citation:
        url = f"https://journals.asm.org/{doi}"
    # Add more conditions for other citation sources
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the 'div' element with class 'affiliations'
    affiliations_div = soup.find('div', class_='Affiliation')
    
    # Check if the 'div' element is found and extract the affiliation
    if affiliations_div:
        affiliation = affiliations_div.text.strip()
    else:
        affiliation = 'N/A'  # Set a default value if affiliation is not found
    
    return affiliation
```

```python
def url_maker(citation, doi):
    url = ''
    # Check the citation source and construct the URL accordingly
    if 'PLOS' in citation:
        url = f"https://journals.plos.org/plosone/article?id={doi}"
    elif 'mBio' in citation:
        url = f"https://journals.asm.org/doi/{doi}"
    elif 'Infect' in citation:
        if 'PMID' in doi:
            pmid = doi.split(': ')[1]
            url = f"https://pubmed.ncbi.nlm.nih.gov/{pmid}"
        else:
            url = f"https://journals.asm.org/doi/{doi}"
    elif 'J. Clin. Microbiol' in citation:
        if 'PMID' in doi:
            pmid = doi.split(': ')[1]
            url = f"https://pubmed.ncbi.nlm.nih.gov/{pmid}"
        else:
            url = f"https://journals.asm.org/doi/{doi}"
    elif 'BioMed Research International' in citation:
        # Add more conditions for other citation sources
        pass
    else:
        print('add more format')
    # Handle unexpected DOI formats
    if not url:
        print(f"Failed to generate URL for DOI: {doi}")

    # Return the constructed URL
    return url
```

```python
import requests
from bs4 import BeautifulSoup

def search_paper_title(title):
    # Construct the search URL with the provided title
    search_url = "https://www.ncbi.nlm.nih.gov/pubmed/?term="
    search_term = title.replace(" ", "+")
    full_url = search_url + search_term

    # Send a GET request to perform the search
    response = requests.get(full_url)

    # Parse the HTML content of the search results page
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the first search result link
    first_result_link = soup.find("a", class_="docsum-title")

    if first_result_link:
        # Extract the URL from the search result
        result_url = 'https://pubmed.ncbi.nlm.nih.gov/' + first_result_link["href"]
        return result_url
    else:
        print("No search results found.")
        return None

```

```python
# Trial:
for index, row in bik.iterrows():
    print(index)
    title = row['Title']
    print('title is: ', title)
    
    # Level 1: Search on NCBI
    result_url = search_paper_title(title)
    if result_url:
        print("Result URL:", result_url)
    else:
        print('No result found on NCBI.')
        
        # Level 2: Use url_maker function
        citation = row['Citation']
        doi = row['DOI']
        url = url_maker(citation, doi)
        if url:
            print("URL generated using url_maker:", url)
        else:
            print("No URL generated using url_maker.")
            
            # Level 3: Handle cases where information is not available
            print("No search results found.")
        
    # Add a separator between iterations for better readability
    print("---")
```

```python
# Create a new column for URLs and update the csv dataset
bik['URL'] = ""

# Example usage:
for index, row in bik.iterrows():
    title = row['Title']
    
    # Level 1: Search on NCBI
    result_url = search_paper_title(title)
    if result_url:
        bik.at[index, 'URL'] = result_url
    else:
        # Level 2: Use url_maker function
        citation = row['Citation']
        doi = row['DOI']
        url = url_maker(citation, doi)
        if url:
            bik.at[index, 'URL'] = url
        else:
            # Level 3: Handle cases where information is not available
            print("No search results found.")
            continue
bik.to_csv('updated_bik.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup

def get_first_result_url(search_term):
    # Website URL
    url = "https://www.researchgate.net/"

    # Send a GET request to the website
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the search box
    search_box = soup.find("input", class_="index-search-field__input")

    # Check if the search box is found
    if search_box:
        print('found')
        # Set the search term in the search box
        search_box["value"] = search_term

        # Submit the form by finding the parent form and calling submit()
        form = search_box.find_parent("form")
        if form:
            response = requests.post(form["action"], data=form.serialize())

            # Parse the search results page
            search_results = BeautifulSoup(response.content, 'html.parser')

            # Find the first result link
            first_result_link = search_results.find("a", class_="nova-legacy-e-link", href=True)

            # Check if the link is found and extract the URL
            if first_result_link:
                result_url = first_result_link["href"]
                return result_url
            else:
                return None
        else:
            return None
    else:
        return None

# Example usage:
search_term = "Chronic Morphine Treatment Attenuates Cell Growth of Human BT474 Breast Cancer Cells by Rearrangement of the ErbB Signalling Network"
result_url = get_first_result_url(search_term)

if result_url:
    print("URL of the first result:", result_url)
else:
    print("No search results found.")
```

# The following part is to gain researchgate link from google search then add it to the original file to create updated_bik_rg.csv

```python
from googlesearch import search

def get_researchgate_link_from_google(paper_title):
    search_query = f"ResearchGate {paper_title}"
    
    try:
        # Perform the Google search and retrieve the first result
        for result in search(search_query, num=1, stop=1, pause=2):
            # Check if the link is from ResearchGate
            if 'researchgate.net' in result:
                return result
            
    except Exception as e:
        print("Error occurred while performing the Google search:", e)
    
    print("No ResearchGate link found.")
    return None

# Example usage:
paper_title = "Chronic Morphine Treatment Attenuates Cell Growth of Human BT474 Breast Cancer Cells by Rearrangement of the ErbB Signalling Network"
researchgate_link = get_researchgate_link_from_google(paper_title)
if researchgate_link:
    print("ResearchGate link found:", researchgate_link)
    # Uncomment the following line to open the link in a web browser
    # webbrowser.open(researchgate_link)
```

# Following part utilize BeautifulSoup on GoogleScholar to search paper title and retrieve the cited number to gain the citation feature.

```python
import requests
from bs4 import BeautifulSoup

def get_citation_count_from_google_scholar(paper_title):
    # Encode the paper title to be used in the search query
    search_query = f"https://scholar.google.com/scholar?q={paper_title.replace(' ', '+')}"
    
    try:
        # Send an HTTP request to Google Scholar
        response = requests.get(search_query)
        response.raise_for_status()
        
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Locate the element containing the citation count
        citation_element = soup.find('a', string=lambda s: s and 'Cited by' in s)
        
        if citation_element:
            # Extract the citation count value
            citation_count = citation_element.text.strip().split('Cited by ')[-1]
            return int(citation_count)
        else:
            print("Citation count not found.")
            return None
        
    except requests.exceptions.RequestException as e:
        print("Error occurred while fetching citation count:", e)
        return None

# Example usage:
paper_title = "Chronic Morphine Treatment Attenuates Cell Growth of Human BT474 Breast Cancer Cells by Rearrangement of the ErbB Signalling Network"
citation_count = get_citation_count_from_google_scholar(paper_title)
if citation_count is not None:
    print("Citation count:", citation_count)
```
