### Using PyPDF2 to skip the first ResearchGate page inside the pdf files

```python
import PyPDF2


def extract_text_from_second_page(pdf_file_path, output_file_path):
    # Open the PDF file
    with open(pdf_file_path, 'rb') as pdf_file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        
        # Check if the PDF has at least two pages
        if len(pdf_reader.pages) >= 2:
            # Initialize an empty string to store the text
            extracted_text = ""
            
            # Start from the second page (index 1) and extract text from each page
            for page in pdf_reader.pages[1:]:
                extracted_text += page.extract_text()
            
            # Save the extracted text to the output file
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(extracted_text)
                
            print(f"Text extracted from the second page and saved to {output_file_path}.")
        else:
            print("PDF does not have at least two pages.")
```

```python
# Example usage
pdf_file_path = '/Users/alberli/Desktop/USC/DSCI 550 DS at Scale/DSCI-550/hw2/trial dataset/LRP-6_is_co-receptor_for_multiple_fibrogenic_pathw.pdf'

# Replace 'output.txt' with the desired output file path
output_file_path = '/Users/alberli/Desktop/USC/DSCI 550 DS at Scale/DSCI-550/hw2/trial.txt'

# Call the function to extract and save the text
extract_text_from_second_page(pdf_file_path, output_file_path)
```

```python
def count_empty_lines_and_add_new_paper(file_path):
    # Read the content of the file
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Function to check if a line is empty
    def is_empty_line(line):
        return line.strip() == ''

    # Initialize variables
    empty_line_count = 0

    for i, line in enumerate(lines):
        if is_empty_line(line):
            empty_line_count += 1
            if empty_line_count > 30:
                # Insert the "New Paper" line at the beginning of the block
                lines.insert(i - empty_line_count + 1, '\\textbf{New Paper}\n')
                # Reset the counter
                empty_line_count = 0
        else:
            # Reset the counter when a non-empty line is encountered
            empty_line_count = 0

    return lines

# Usage example:
file_path = 'result_USC.txt'
modified_lines = count_empty_lines_and_add_new_paper(file_path)

# Save the modified lines back to the file
with open('trial1.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(''.join(modified_lines))
```

```python
# Function to remove lines with only non-letter characters or 
# less than 3 letters
def trimmer(text_lines):
    cleaned_lines = [line for line in text_lines if 
                     any(char.isalpha() for char in line) and 
                     len([char for char in line if char.isalpha()]) >= 3]
    return cleaned_lines

# Read the content of the file
with open('trial2_5.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Remove lines with only non-letter characters or less than 3 letters
cleaned_lines = trimmer(lines)

# Join the lines back into a single string
cleaned_text = ''.join(cleaned_lines)

# Save the cleaned text back to the "trial2.txt" file
with open('trial3.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(cleaned_text)
```

```python
# Function to adjust lines with consecutive "New Paper" phrases and add new lines above and below
def line_adjustment(text_lines):
    cleaned_lines = []
    new_paper_count = 0
    for line in text_lines:
        if "\\textbf{New Paper}" in line:
            new_paper_count += 1
            if new_paper_count < 2:
                cleaned_lines.append("\n" + line)
        else:
            new_paper_count = 0
            cleaned_lines.append(line)
    return cleaned_lines

# Read the content of the file
with open('trial3.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Remove lines with only non-letter characters or less than 3 letters
cleaned_lines = trimmer(lines)

# Adjust lines with consecutive "New Paper" phrases and add new lines above and below
cleaned_lines = line_adjustment(cleaned_lines)

# Join the lines back into a single string
cleaned_text = ''.join(cleaned_lines)

# Save the cleaned text back to the "trial2.txt" file
with open('trial3.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(cleaned_text)
```

### Now after many times of cleaning, the three lines under the New Paper is highly likely to be the title itself. Thus, We'll make the necessary latex format based on it.

```python
def final_latex(text_lines, substrings_to_remove_after_new_paper):
    cleaned_lines = []
    new_paper_count = 0

    for line in text_lines:
        if "\\textbf{New Paper}" in line:
            new_paper_count += 1
            if new_paper_count > 1:
                cleaned_lines.append("\\end{document}\n")  # Close the previous content section
            cleaned_lines.append("\n\\documentclass{article}\n\\begin{document}\n")
            cleaned_lines.append("\\section*{}\n")
        cleaned_lines.append(line)

    # Add the closing content section for the last paper
    if new_paper_count > 1:
        cleaned_lines.append("\\end{document}\n")

    return cleaned_lines

# Rest of the code remains the same...
```

```python
# Read the content of the "trial4_5.txt" file
with open('trial4_5.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Substrings to remove after the "\\textbf{New Paper}" line
# based on the excel output above
substrings_to_remove_after_new_paper = ["research", "reload", 
                                        "Retraction", "biomed", 
                                        "cytokine", "journal", 
                                        "deb", "pbio"]

# Clean the content and apply LaTeX formatting
cleaned_lines = final_latex(lines, substrings_to_remove_after_new_paper)

# Join the lines back into a single string
cleaned_text = ''.join(cleaned_lines)

# Save the cleaned text into a new file named "final_output.tex"
with open('trial5.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(cleaned_text)
```

```python
def content_function(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        add_section = False
        count = 0
        for line in infile:
            outfile.write(line)
            if '\\textbf{New Paper}' in line:
                add_section = True
                count = 0
            if add_section:
                count += 1
                if count == 3:
                    outfile.write('\\section*{}\n')
                    add_section = False
                    count = 0

if __name__ == '__main__':
    input_file = 'trial5.txt'
    output_file = 'final_trial.txt'
    content_function(input_file, output_file)
```
