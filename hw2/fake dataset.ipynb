```python
import pandas as pd
pd.set_option('display.max_columns',None)
df=pd.read_csv('third_dataset.csv')
```

## Fake name

```python
lst_author=df['Author count'].to_list()
```

```python
import random
from faker import Faker
fake = Faker()
lst_name=[]
for i in range(500):
    num=random.sample(lst_author,k=1)[0]
    name=''
    for _ in range(num):
        name+=fake.name()+', '
    lst_name.append(name.rstrip(', '))
```

```python
df_new=pd.DataFrame(lst_name)
df_new.columns=['Authors']
```

## Fake title/ affiliation

```python
lst=[]
df=df[~df['Title'].isnull()]
for i in df['Title']:
    lst.append(i)
lst_t=list(set(lst))  #list of title

lst_title=[]    #fake title
for i in range(500):
    t = random.choice(lst_t)
    lst_title.append(t)
df_new['Title']=pd.DataFrame(lst_title)
```

```python
#Title (generated by GPT2)
import numpy as np
import os
path='fake_content_final_trial/'  
files= os.listdir(path)
count=0
dl= []
for file in files: 
    dl.append(pd.read_csv(path+file))
df0=pd.concat(dl)
df0=df0[:250]      #Trained with Latex
df0.drop("Unnamed: 0", axis=1, inplace=True)


df_temp=pd.read_csv('fake_generated_merged.csv')
df_temp=df_temp[:250]    #Trained without Latex


bik=pd.concat([df0, df_temp])
bik=bik.reset_index(drop=True)  


title=[]
content=[]
for i in bik['Content']:
    title.append(i.split('\n')[0][:120])  # control the length under the academically recommended range
    content.append(i.split('\n')[1:])
```

```python
df_new['Title']=pd.DataFrame(title)
```

```python
#Affiliation (random sample)
lst=[]
df=df[~df['university/institution'].isnull()]
for i in df['university/institution']:
    lst.append(i)
lst_af=list(set(lst))  #list of affiliation

lst_faf=[]    #fake affiliation
for i in range(500):
    af = random.choice(lst_af)
    lst_faf.append(af)
#print(lst_faf)
df_new['university/institution']=pd.DataFrame(lst_faf)
```

## Location

```python
lst=[]
for i in range(len(df_new)):
    if df_new['university/institution'][i] in df['university/institution'].to_list():
        index=df[df['university/institution']==df_new['university/institution'][i]].index.to_list()[0]
        location=df['Location'][index]
        lst.append(location)
df_new['Location']=pd.DataFrame(lst)
```

## Other features

```python
def dataframe(str):
    lst=[]
    for i in df[str]:
        lst.append(i)
    lst_temp=[]
    for i in range(500):
        temp = random.choice(lst)
        lst_temp.append(temp)
    df_new[str]=pd.DataFrame(lst_temp)
    return df_new
```

```python
columns=df.columns.to_list()[4:-5]+df.columns.to_list()[-3:]
```

```python
for i in columns:
    dataframe(i)
```

```python
# Correct First Author
lst=[]
for i in df_new['Authors']:
    lst.append(i.split(',')[0])
df_new['First Author']=pd.DataFrame(lst)
```
