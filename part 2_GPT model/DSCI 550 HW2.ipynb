### Using PyPDF2 to skip the first ResearchGate page inside the pdf files

```python
import PyPDF2


def extract_text_from_second_page(pdf_file_path, output_file_path):
    # Open the PDF file
    with open(pdf_file_path, 'rb') as pdf_file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        
        # Check if the PDF has at least two pages
        if len(pdf_reader.pages) >= 2:
            # Initialize an empty string to store the text
            extracted_text = ""
            
            # Start from the second page (index 1) and extract text from each page
            for page in pdf_reader.pages[1:]:
                extracted_text += page.extract_text()
            
            # Save the extracted text to the output file
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(extracted_text)
                
            print(f"Text extracted from the second page and saved to {output_file_path}.")
        else:
            print("PDF does not have at least two pages.")
```

```python
def count_empty_lines_and_add_new_paper(file_path):
    # Read the content of the file
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Function to check if a line is empty
    def is_empty_line(line):
        return line.strip() == ''

    # Initialize variables
    empty_line_count = 0

    for i, line in enumerate(lines):
        if is_empty_line(line):
            empty_line_count += 1
            if empty_line_count > 30:
                # Insert the "New Paper" line at the beginning of the block
                lines.insert(i - empty_line_count + 1, '\\textbf{New Paper}\n')
                # Reset the counter
                empty_line_count = 0
        else:
            # Reset the counter when a non-empty line is encountered
            empty_line_count = 0

    return lines

# Usage example:
file_path = 'result_USC.txt'
modified_lines = count_empty_lines_and_add_new_paper(file_path)

# Save the modified lines back to the file
with open('trial1.txt', 'w', encoding='utf-8') as output_file:
    output_file.write(''.join(modified_lines))
```

```python
def final_latex(text_lines, substrings_to_remove_after_new_paper):
    cleaned_lines = []
    new_paper_count = 0

    for line in text_lines:
        if "\\textbf{New Paper}" in line:
            new_paper_count += 1
            if new_paper_count > 1:
                cleaned_lines.append("\\end{document}\n")  # Close the previous content section
            cleaned_lines.append("\n\\documentclass{article}\n\\begin{document}\n")
            cleaned_lines.append("\\section*{}\n")
        cleaned_lines.append(line)

    # Add the closing content section for the last paper
    if new_paper_count > 1:
        cleaned_lines.append("\\end{document}\n")

    return cleaned_lines

# Rest of the code remains the same...
```
